## **Conversão de tipos**

* Ao carregar os dados, o notebook interpretou a maioria deles como _float64_, seguido de _object_ e _int64_, o que não está completamente correto. 
* Observa-se no **_"Dicionário de Dados"_** que não há entradas de valores com pontos flutuantes, também há a questão de haver datas, mas o tipo _datetime_ não ter aparecido no resumo das informações, o que dá a entender que foram interpretados como _object_.

### **float64 → Int64**

* Como mencionado, não há entradas de ponto flutuante na tabela, o que pode ter ocorrido é um problema que o _int64_ do _Numpy_ tem. Ele não aceita valores nulos (NaN), então ao carregar a tabela o _Pandas_ leu um campo NaN em uma coluna _int64_ e a converteu para o tipo _float64_, que tem suporte para NaN.
* O _Pandas_ tem um tipo especial para resolver esse exato problema que é o _Int64_ (**_Integer extension type_**), sua criação é justamente para conseguir interpretar valores nulos.

```python
#conversão de float sem casa decimais para Int64, object para string e manter o NaN como <NA> (versão de nulidade do Pandas)
df = df.convert_dtypes()
```

* O tipo Float é atualizado para Integer e o tipo Object é atualizado para String.

```python
df.info()
```

* Os dados deixam de ter os 0's após a vírgula e se tornam números inteiros.

```python
df.head(3)
```

### **string → datetime**

* Por fim tem a conversão das datas que inicialmente foram lidas como object e agora estão como strings.

```python
#exemplo de uma coluna com datas sendo lida como string
print(df['DT_NASC'].dtype)
```

* A base de dados segue o padrão de começar a maioria dos nomes das colunas de data com 'DT...', então há como filtrar grande parte das colunas e mudar o tipo sem precisar as listar uma a uma.

```python
#filtro que retornará todas colunas que começam com DT
df.filter(regex = r'^DT').head(3)
```

* As outras colunas de data fazem alusão a datas de quando os pacientes tomaram vacina, ou seja, o restante das colunas começam com 'DOS...'.

```python
df.filter(regex = r'^DOS').head(3)
```

* As modificações para as datas são: exclusão dos horários e modificação para datetime.

```python
#seleciona as colunas que começam com 'DT' e 'DOS'.
columns_dt = df.filter(regex=r'^DT').columns
columns_dos = df.filter(regex=r'^DOS').columns

#exclui os horários das colunas 'DT...'.
df[columns_dt] = df[columns_dt].apply(
    lambda col: pd.to_datetime(col, errors='coerce').dt.date
)

#atualiza as colunas que começam com 'DT' para o tipo datetime
df[columns_dt] = df[columns_dt].apply(pd.to_datetime, errors='coerce')

#atualiza as colunas que começam com 'DOS' para o tipo datetime
df[columns_dos] = df[columns_dos].apply(pd.to_datetime, format='%d/%m/%Y', errors='coerce')

```

* É fácil corrigir essa observação, pois basta avaliar as variáveis DT_NASC (data de nascimento), DT_SIN_PRI (data dos primeiros sintomas) e TP_IDADE (tipo idade).

```python
df.loc[31746452307100, ['DT_SIN_PRI', 'DT_NASC', 'TP_IDADE', 'NU_IDADE_N']]
```

* Nessa observação vemos que a idade por algum motivo foi classificada como do Tipo 2 (mês) sendo que o correto, de acordo com o dicionário de dados, seria ser classificada como do Tipo 1 (dias), como é algo simples de ser corrigido não optarei pela exclusão da linha, apenas farei algumas modificações nas colunas NU_IDADE_N e TP_IDADE.
* No dicionário o cálculo da idade é baseado em _data dos primeiros sintomas_ - _data do nascimento_, e por esse cálculo, nessa observação específica, fazer alusão a um recém-nascido, com menos de um mês de vida, o tipo de idade dele deve ser classificado como 1(dias).

```python
#mudança da idade para 14 (dias) e o tipo para 1.
df.loc[31746452307100, ['NU_IDADE_N', 'TP_IDADE']] = [14, 1]
```

```python
df.loc[31746452307100, ['DT_SIN_PRI', 'DT_NASC', 'TP_IDADE', 'NU_IDADE_N']]
```

* Por fim, apenas para garantir, conferirei inconsistências nas idades registradas com base no tipo nos quais elas foram classificadas.

```python
#filtros para separar as idades de acordo com o tipo (dias, meses, anos)
filter_day = df['TP_IDADE'] == 1
filter_month = df['TP_IDADE'] == 2
filter_year = df['TP_IDADE'] == 3

#seleção das linhas que pertencem a cada categoria 
df_days = df.loc[filter_day].copy()
df_months = df.loc[filter_month].copy()
df_years = df.loc[filter_year].copy()

# subtração da data dos primeiros sintomas menos a data que contém o nascimento em cada observação 
df_days['IDADE_CALCULADA_DIA'] = (
    (df_days['DT_SIN_PRI'] - df_days['DT_NASC']).dt.days
)

df_months['IDADE_CALCULADA_MES'] = (
    (df_months['DT_SIN_PRI'] - df_months['DT_NASC']).dt.days // 30
)

df_years['IDADE_CALCULADA_ANO'] = (
    (df_years['DT_SIN_PRI'] - df_years['DT_NASC']).dt.days // 365
)

# criação das variáveis para visualizar comparações 

difference_days = df_days.loc[
    df_days['IDADE_CALCULADA_DIA'] != df_days['NU_IDADE_N'],
    ['DT_NASC', 'DT_SIN_PRI', 'NU_IDADE_N', 'IDADE_CALCULADA_DIA']
]

difference_months = df_months.loc[
    df_months['IDADE_CALCULADA_MES'] != df_months['NU_IDADE_N'],
    ['DT_NASC', 'DT_SIN_PRI', 'NU_IDADE_N', 'IDADE_CALCULADA_MES']
]

difference_years = df_years.loc[
    df_years['IDADE_CALCULADA_ANO'] != df_years['NU_IDADE_N'],
    ['DT_NASC', 'DT_SIN_PRI', 'NU_IDADE_N', 'IDADE_CALCULADA_ANO']
]
```

* Apesar de haver discrepâncias nos tipos 2 e 3 (mês, ano), elas são de apenas um número, ou seja, não são tão "fora da curva". Entretanto, quando se trata do tipo 1 (dia), os valores tendem a ser muito diferentes entre si, então vou focar apenas nesse último e fazer as modificações necessárias. Há três tipos de atualizações que precisam ser feitas:
    * 1° - idades que pertencem ao tipo 1, mas por algum motivo o sistema calculou erroneamente seus valores;
    * 2° - idades que partencem ao tipo 2, porém estão registrados como do tipo 1 (campos que possuem valores acima de 29 dias e abaixo de 364);
    * 3° - idades que pertencem ao tipo 3, porém estão registrados como do tipo 1 (campos que possuem valores acima de 364 dias).

```python
# correção das idades que pertencem ao tipo 1, mas que apresentam erros no registro
real_age = (
    (df['DT_SIN_PRI'] - df['DT_NASC']).dt.days
)

filtro = (
    (df['TP_IDADE'] == 1) &
    (real_age < 30) &
    (df['NU_IDADE_N'] != real_age)
)

df.loc[filtro, 'NU_IDADE_N'] = (
    real_age[filtro]
).astype(int)
```

```python
df.loc[31738260806730, ['NU_IDADE_N']]
```

```python
# correção das idade que estão registradas como tipo 1, mas pertencem ao tipo 2
filtro = (
    (df['TP_IDADE'] == 1) &
    (
        ((df['DT_SIN_PRI'] - df['DT_NASC']).dt.days > 29) &
        ((df['DT_SIN_PRI'] - df['DT_NASC']).dt.days < 365)
    )
)

# Atualizar TP_IDADE
df.loc[filtro, 'TP_IDADE'] = 2

# Atualizar NU_IDADE_N
df.loc[filtro, 'NU_IDADE_N'] = (
    np.ceil(
        ((df.loc[filtro, 'DT_SIN_PRI'] -
          df.loc[filtro, 'DT_NASC']).dt.days) / 30
    )
).astype(int)
```

```python
df.loc[31753292927733, ['TP_IDADE', 'NU_IDADE_N']]
```

```python
# correção das idade que estão registradas como tipo 1, mas pertencem ao tipo 3
filtro = (
    (df['TP_IDADE'] == 1) &
    (
        (df['DT_SIN_PRI'] - df['DT_NASC']).dt.days > 364
    )
)

# Atualizar TP_IDADE
df.loc[filtro, 'TP_IDADE'] = 3

# Atualizar NU_IDADE_N
df.loc[filtro, 'NU_IDADE_N'] = (
    np.ceil(
        ((df.loc[filtro, 'DT_SIN_PRI'] -
          df.loc[filtro, 'DT_NASC']).dt.days) // 365
    )
).astype(int)
```
